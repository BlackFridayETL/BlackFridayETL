```python
import pandas as pd
from sqlalchemy import create_engine
from config import un, pw, uri, port
import pymysql
pymysql.install_as_MySQLdb()
```

```python
csv_file = "blackfriday_deals.csv"
blackfriday_deals_df = pd.read_csv(csv_file)
blackfriday_deals_df.head()
```

```python
csv_file2 = "Incident.csv"
incidents_df = pd.read_csv(csv_file2)
incidents_df.head()
```

```python
blackfriday_deals = blackfriday_deals_df[['Store', 'Category', 'Description', 'Year']].copy()
blackfriday_deals
```

```python
incidents = incidents_df[['Store', 'Place', 'Deaths', 'Injuries', 'Year', 'Cause']].copy()
incidents_11 = incidents.loc[incidents['Year'] == 2011, :]
incidents_11.head()
```

```python
rds_connection_string = f"{un}:{pw}@{uri}:{port}/blackfriday_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
blackfriday_deals.to_sql(name='deals', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from deals', con=engine)
```

```python
incidents_11.to_sql(name='incidents', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from incidents', con=engine).head()
```
